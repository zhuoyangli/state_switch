```python
# 1. extract full voi of functional data: schaefer-n200-net7

#note that sub-008 has started 1tr early than supposed, so mannually removed first tr, and duplicated last tr at the end
import xianfunc as xf
data_dir = '../../data/preproced/fsl_complete/'
sub_list = ['sub-065']#xf.grab_fname(data_dir,'sub*')
task_list = ['carver','ntf']
maskn,netn = 200,7
parcel = 'n200_net7'

for task in task_list:
    print(task)
    for sub in sub_list:
        print(sub)
        func_file = data_dir+sub+'/'+sub+'_'+task+'.nii.gz'
        out = '../../analysis/results/1_data/voi/'+task+'_'+parcel+'/'+sub+'_'+task+'_voi_n'+str(maskn)+'_net'+str(netn)+'.csv'
        if xf.file_exist(out)==False:
            voi = xf.extract_voi(func_file,maskn=maskn,netn=netn,outfile=out,to_csv=True)
print('completed')
```

```python
# 2. extract full voi of functional data: schaefer-n500-net17

#note that sub-008 has started 1tr early than supposed, so mannually removed first tr, and duplicated last tr at the end
import xianfunc as xf
data_dir = '../../data/preproced/fsl_complete/'
sub_list =['sub-065']# xf.grab_fname(data_dir,'sub*')
task_list = ['carver','ntf']
maskn,netn = 500,17
parcel = 'n500'

for task in task_list:
    for sub in sub_list:
        func_file = data_dir+sub+'/'+sub+'_'+task+'.nii.gz'
        out = '../../analysis/results/1_data/voi/'+task+'_'+parcel+'/'+sub+'_'+task+'_voi_n'+str(maskn)+'.csv'
        if xf.file_exist(out)==False:
            print(task,sub)
            voi = xf.extract_voi(func_file,maskn=maskn,netn=netn,outfile=out,to_csv=True)
print('completed')
```

```python
# 3. save all .nii file to .hdf5 file for faster loading (not sure will use, but just in case)

import xianfunc as xf
data_dir = '../../data/preproced/fsl_complete/'
sub_list =['sub-065']# xf.grab_fname(data_dir,'sub*')
task_list = ['carver','ntf']

for task in task_list:
    print(task)
    for sub in sub_list:
        print(sub)
        func_file = data_dir+sub+'/'+sub+'_'+task+'.nii.gz'
        xf.make_hdf5(func_file)
print('completed')
```

```python
# 4. extract mvp for each tr given each roi mask

import xianfunc as xf
data_dir = '../../data/preproced/fsl_complete/'
sub_list = xf.grab_fname(data_dir,'sub*') #['sub-065']#
task_list = ['carver','ntf']
roi_list = ['NAc'] #['hipp_a','hipp_p','hipp','amy','strm']#['acc','amyg','hc','mpfc_v','mpfc_d','pcc','a1','hc','pmc','v1','dmn']

#note that sub-008 need to shift 1tr later in analysis for MVP
for task in task_list:
    print(task)
    for roi in roi_list:
        print(roi)
        for sub in sub_list:
            print(sub)
            func_file = data_dir+sub+'/'+sub+'_'+task+'.nii.gz'
            out = '../../analysis/results/1_data/mvp/'
            out_f = out+sub+'_'+task+'_'+roi+'_mvp.csv'
            if xf.file_exist(out_f)==False:
                mvp_ls = xf.extract_mvp_raw(func_file,'../masks/'+roi+'.nii',outpath=out)
print('completed')
```

```python
import xianfunc as xf
task_list = ['carver','ntf']
cond_list = ['continuous','intact_pause','intact_tom','scram_pause']
carver_exclude = ['sub-'+i for i in ['016','054','056','064','066','076','087']]
ntf_exclude = ['sub-'+i for i in ['017','044','056','087']]

#get the cond info for each preprocessed subject
sub_exist = xf.grab_all('../../data/preproced/fsl_complete/','*')
subinfo = xf.read_data('fsl_preproc.xlsx',[1,2])
subinfo['subj'] = xf.replace_str(subinfo['subj'],'int','sub-')
kick = xf.merge_lists_in_list([[i for i in subinfo['subj'] if i not in sub_exist],carver_exclude,ntf_exclude])
subinfo = xf.dict_exclude_rows_multi(subinfo,'subj',kick)
print(len(subinfo['subj']))
subinfo['subj']
```

```python
# 1) Schaefer n200 DMN (get the dmn regions)
# 2) DMN masked mvp (get the average VOI)
# 3) pearson correlation between the two: just check out several subjects--make sure r is high

import xianfunc as xf
lab_file = '../masks/use_labels/schaefer2018_n200_label.csv'
label = xf.read_csv(lab_file,asnum=False)
label = [i[0] for i in label]
inds = xf.get_indices(search_list=label,search_item='Default',find=True)
dmns = [label[ind] for ind in inds]

voi_folder = '../results/1_data/voi/carver_n200_net7/'
mvp_folder = '../results/1_data/mvp/'
for subj in subinfo['subj'][0:10]:
    #this is dmn VOI from schaefer n200-net7 46 regions with 'Default' in its label 
    voi_file = voi_folder+xf.grab_all(voi_folder,subj+'*')[0]
    voi = xf.load_voi(voi_file)
    voi_dmn = [[j for j in voi[str(i)]] for i in inds]
    voi_dmn_avg = xf.avg(voi_dmn,mute=True)
    
    #this is dmn VOI from dmn-masked mvp average
    mvp_file = mvp_folder+xf.grab_all(mvp_folder,subj+'_carver_dmn*')[0]
    mvp = xf.read_csv(mvp_file,skip_row1=True)
    mvp_dmn_avg = xf.avg(mvp,mute=True)   
    
    r,p = xf.get_corr(voi_dmn_avg,mvp_dmn_avg)
    print(r)
```
